````python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
````

````python
plt.xlim(0,2*np.pi)
plt.ylim(-2,2)
plt.xlabel('x')
plt.ylabel('f(x)',fontsize='xx-large')

myxspace = np.linspace(0, 2*np.pi, 100)
A = np.genfromtxt('T04_plot_data.csv', delimiter=',')
B = np.asarray([0, (1/4)*np.pi, (1/2)*np.pi, (3/4)*np.pi, np.pi, (5/4)*np.pi, (3/2)*np.pi, (7/4)*np.pi])
def f(x):
    return ((0.1)*x*(x - 3)*(x - 5))
def g(x):
    return np.sin(x)

plt.plot(myxspace,g(myxspace),'r',label=r"$sin(x)$")
plt.plot(myxspace,f(myxspace),'g:',label=r"$0.1x(x-3)(x-5)$")
plt.plot(A[:,0], A[:,1], 'k--', label = 'data')
plt.plot([0, 2*np.pi], [-1, -1], 'b')
plt.plot(B, g(B), 'b+')
plt.legend(loc = 'upper left')
plt.savefig('T04_plot.pdf')
````
